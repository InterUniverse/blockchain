# Transform Ethereum raw transactions

### Read the eth_transactions.json files

```python
import os
import json

import pandas as pd
import numpy as np
```

```python
PATH = '../data/'
```

```python
pd.set_option('display.max_columns', 100, 'display.max_rows', 100)
pd.set_option('display.max_colwidth', 100)
```

```python
file_dir = os.listdir(PATH)
file_list = [file for file in file_dir if file.endswith('transactions.json')]
```

```python
%%time
raw_json_list = [json.load(open(f'{PATH}{file}')) for file in file_list]
```

```python
%%time
df_raw = pd.concat([pd.DataFrame(file) for file in raw_json_list])
```

```python
df_raw.shape
```

```python
df_raw.info()
```

```python
df = df_raw.copy()
```

```python
col_list = ['blockNumber','chainId','gas','gasPrice','maxFeePerGas','maxPriorityFeePerGas','nonce','transactionIndex','type','v','value']
```

```python
df_raw.sample(5)
```

```python
%%time
for col in col_list:
    df[col] = df[col].apply(lambda x: int(x, base=16) if type(x) is str else x)
```
